* parse_date = True, informa que o banco de dados possui uma coluna do data

### Visualização de dados

```python
def plot_bar(variavel,
             data,
             figsize = (10, 5),
             bins = 30):
    values = data[variavel].value_counts()
    x = values.index
    plt.figure(figsize = figsize)
    plt.bar(x, values)
    plt.ylabel('Frequência')
    plt.xticks(x, values.index.values)
    plt.title(variavel)
    plt.grid()
    plt.show()
```

```python
def plot_histograma(variavel, 
               data, 
               figsize = (10, 5), bins = 30,
               ylabel = 'Frequência',
              ):
    "Gera gráfico do tipo histograma para uma variável do dataset."
    plt.figure(figsize = figsize)
    plt.hist(data[variavel], bins = bins)
    plt.xlabel(variavel.title())
    plt.ylabel(ylabel)
    plt.title(f'Histograma - {variavel}')
    plt.grid()
    plt.show()
```

```python
variaveis_numericas = df.select_dtypes(include = ['int64', 'float64']).columns.to_list()[1:]
for variavel in variaveis_numericas:
    print(plot_histograma(variavel, df))
```

```python
variaveis = ['Survived', 'Sex', 'Embarked']
for variavel in variaveis:
    plot_bar(variavel, df)
```
